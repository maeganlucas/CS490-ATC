```python
import pydub
import numpy as np
import requests
import io
from matplotlib import pyplot as plt
```

```python
# https://stackoverflow.com/a/34159184


stream_url = 'http://d.liveatc.net/kdab_del_gnd'

r = requests.get(stream_url, stream=True)

arr = []

try:
    for block in r.iter_content(10240):
        # print(block)
        # print('\n\n')
        # s = io.BytesIO(block)

        a = pydub.AudioSegment(block, sample_width=2, frame_rate=11025, channels=1)

        arr += np.array(a.get_array_of_samples())

        

except KeyboardInterrupt:
    pass
```

```python

plt.plot(list(range(0,len(arr))), arr)
plt.show()
```

```python
with open('stream.wav', 'rb') as f:
    a = pydub.AudioSegment.from_mp3(f)
    y = np.array(a.get_array_of_samples())
    if a.channels == 2:
        y = y.reshape((-1, 2))

f.close()
```
